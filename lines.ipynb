```python
# 1. Draw a line in a diagram from position (0,0) to position (6,250):

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# 2. Plotting Without Line

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints,'o')
plt.show()
```

```python
# 3. Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# 4. Draw a the points in a diagram from position (1, 3)  (2, 8)  (6, 1)  (8, 10):

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints,'o')
plt.show()
```

```python
# 5. If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3

# (0,3) (1,8) (2,1) (3,10) (4,5) (5,7)

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

```python
# marker - mark each points

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([3, 8, 4, 10,7])
ypoints = np.array([3, 3, 1, 15,4])

plt.plot(ypoints,xpoints, marker = 's')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])
xpoints = np.array([4, 8, 7, 6])


plt.plot(xpoints,ypoints,color='red',linestyle = 'dashed', marker ='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])
xpoints = np.array([4, 8, 7, 6])


plt.plot(xpoints,ypoints,color='red',linestyle = 'dotted', marker ='o' , markersize=20)
plt.show()
```

```python
#marker color (mec='g')

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])
xpoints = np.array([4, 8, 7, 6])


plt.plot(xpoints,ypoints,color='red',linestyle = 'dotted', marker ='o' , markersize=20 , mec='g')
plt.show()
```

```python
# markerfacecolor or the shorter mfc to set the color inside the edge of the markers:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', color='black', markersize = 20, linestyle='dashed', markerfacecolor = 'r', linewidth = '10')
plt.show()
```

```python
#with formatstring

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o-.g')
plt.show()
```
